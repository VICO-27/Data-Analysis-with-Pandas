# Handling Missing Data

Real-world datasets often contain missing values.

We handle them as follows:

Age → Replace missing values with the **median age**  
Embarked → Replace missing values with the **mode (most common value)**  
Cabin → Drop the column due to excessive missing values

```python
# 1. Fill missing Age values with the median
df["Age"] = df["Age"].fillna(df["Age"].median())

# 2. Fill missing Embarked values with the mode (most common value)
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# 3. Drop the Cabin column (reassignment is cleaner than inplace)
df = df.drop(columns=["Cabin"])
```

# Removing Duplicate Rows

Duplicate rows can distort analysis results.

We remove them using `.drop_duplicates()` to ensure data integrity.

```python
df.drop_duplicates(inplace=True)
```

# Survival Rate by Gender

We calculate survival rates grouped by gender using **groupby()**.

This reveals whether males or females were more likely to survive.

```python
df.groupby("Sex")["Survived"].mean()
```

# Survival Rate by Passenger Class

Passenger class played an important role in survival probability.

We analyze survival rates across:
- First class
- Second class
- Third class

```python
df.groupby("Pclass")["Survived"].mean()
```

# Average Age by Class

We compute the average age of passengers within each class to understand demographic differences.

```python
df.groupby("Pclass")["Age"].mean()
```

# Age Group Analysis

We divide passengers into age groups:
- Child
- Teen
- Young Adult
- Adult
- Senior

Then we calculate survival rates for each group.

```python


bins = [0, 12, 18, 35, 60, 100]
labels = ["Child", "Teen", "Young Adult", "Adult", "Senior"]

# Create the AgeGroup column
df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)

df.groupby("AgeGroup", observed=False)["Survived"].mean()
```

# Filtering Female Survivors

We filter passengers who satisfy two conditions:
- Gender = Female
- Survived = 1

```python
female_survivors = df[(df["Sex"] == "female") & (df["Survived"] == 1)]

female_survivors.head()
```

# Children Who Survived

Children are defined here as passengers with **Age < 12**.

We filter the dataset to identify surviving children.


```python
children_survived = df[(df["Age"] < 12) & (df["Survived"] == 1)]

children_survived.head()
```

# First Class Survivors

First class passengers had better access to lifeboats.

We filter passengers who:
- Were in first class
- Survived the disaster

```python
first_class_survivors = df[(df["Pclass"] == 1) & (df["Survived"] == 1)]

first_class_survivors.head()
```

# Key Insights from the Titanic Dataset

### 1. Who was more likely to survive?
Female passengers had a significantly higher survival rate than males.  
This reflects the historical evacuation policy **“Women and children first.”**

### 2. Did passenger class affect survival?
Yes. Passengers in **First Class** had the highest survival probability, while **Third Class** passengers had the lowest survival rates.

### 3. Were children prioritized?
Yes. Children had higher survival rates compared to many adult passengers, suggesting rescue priority during evacuation.

### 4. What combination had the highest survival rate?
The group with the highest survival probability was:

**Female + First Class + Young Age**

These passengers had the best access to lifeboats and evacuation priority.
